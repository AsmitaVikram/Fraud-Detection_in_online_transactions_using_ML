 Importing the dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset to a Panda Dataframe
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
#first 5 row of the dataset
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
#important information about the dataset
credit_card_data.info()
```

```python
#checking the missing values in each column in dataset
credit_card_data.isnull().sum()
```

Distribution of legit transaction and fraudulent transactions


```python
credit_card_data['Class'].value_counts()
```

This Dataset is highly unbalanced

0--> Normal Tansaction

1--> Fraudulent Transaction

```python
#Separating the data from analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Statistical measures of the data

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#comapare the values for both transactions
credit_card_data.groupby('Class').mean()
```

DEALING WITH UNBALANCED DATASET USING UNDER-SAMPLING

Build a sample dataset containing similar distribution of normal transactions and fraudulent transaction

Number of fraudulent transaction --> 492

```python
legit_sample = legit.sample(n=492)
```

Concatinating two DataFrames  

```python
new_dataset = pd.concat([legit_sample,fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into Features and targets

```python
X = new_dataset.drop(columns ='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into Training data and Testing Data


```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2, stratify=Y,random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Using Logistic Regression Model

```python
model = LogisticRegression()
```

```python
#training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
#accuracy on training data
X_train_predication = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_predication, Y_train)
```

```python
print('Accuracy on Training data :', training_data_accuracy)
```

```python
#accuracy on the test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on test Data: ', test_data_accuracy)
```
